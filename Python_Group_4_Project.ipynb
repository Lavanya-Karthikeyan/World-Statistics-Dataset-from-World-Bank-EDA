#### <b>Cleaning the Data</b>

```python
# Dropping duplicates to remove redundant data
print("The shape of the data before dropping duplicate records is ", df.shape)
df = df.drop_duplicates()
print("The shape of the data after dropping duplicate records is ", df.shape)

# Converting all string column to lower case and stripping an leading or trailing spaces
df['Country Name'] = df['Country Name'].str.lower()
df['Country Code'] = df['Country Code'].str.lower()
df['Country Name'] = df['Country Name'].str.strip()
df['Country Code'] = df['Country Code'].str.strip()

# Removing 'Country Code' as we are using the 'Country Name' to uniquely identify each country
del df["Country Code"]


# Adding the Urban Population
df["Urban Population"] = df["Population, total"]-df["Rural population"]

# Displaying the data after cleaning
display(df.head(5))

# GDP column creation
df["GDP"] = df["GDP per capita (current US$)"] * df["Population, total"]

```

#### <b>Getting List of Valid Countries</b>

```python
# The column "Country Name" contains invalid country names. Therefore, we are using the pycountry libray to identify which of the countries in the dataset are valid countries.

# The funtion below takes a parameter of list of all the country values present in the "Country Name" column of the dataset. Furthermore, it returns a list (subset) 
# of those country names which are valid.

def get_valid_countries(list_of_countries):
    
    valid_countries = []
    for c in pycountry.countries:
        for country in list_of_countries:
            if((str(c.name.lower()) == country.lower())):
                valid_countries.append(country.lower())
            
    return(valid_countries)

def get_lat_long(list_of_countries,df):
    
    # Initialize Nominatim API
    from geopy.geocoders import Nominatim
    geolocator = Nominatim(user_agent="whiehr")
    
    country = []
    lat = []
    long = []

    for i in list_of_countries:
        location = geolocator.geocode(i)
        lat.append(location.latitude)
        long.append(location.longitude)
        country.append(i)
    
    print(location)
    print(lat)
    print(long)

    geo = pd.DataFrame(data =[country,lat,long],columns=["Country Name","latitude","longitude"])
    df = df.merge(geo, on =["Country Name"],how="left")
    return df
        
        
```

```python
# Here we are first generating the sibset of countries in the data which are valid country names.

print("The shape of the data before removing invalid country names:", df.shape)

list_of_countries = list(df["Country Name"].unique())
print("The total number of unique \'Country Name\' values in the data (includes invalid names)", len(list_of_countries))

valid_country_list = get_valid_countries(list_of_countries)
print("The total number of unique valid \'Country Name\' values in the data",len(valid_country_list))

# Once we have the list of valid countries we are retaining only that data with valid country names in the main dataframe.

df = df.loc[df["Country Name"].isin(valid_country_list)]
print("The shape of the data with only valid country names:",df.shape)


# df = get_lat_long(valid_country_list,df)
# print("done")

```

#### <b>Calculate the correlations between the variables</b> ####

```python
# Writing a function to find the correlation of the numerical columns

def find_correlation(df1):
    num_cols = list(df1.columns)[2:]
    df1 = df1.fillna(0)
    cdf = pd.DataFrame(columns=["Attribute_1","Attribute_2","Correlation_Co-efficient"])
    for i in num_cols:
        for j in num_cols:
            if(i!=j):
                x = np.corrcoef(df1[i].to_numpy(), df1[j].to_numpy())[0,1]
                new_row = pd.DataFrame({"Attribute_1":i,"Attribute_2":j,"Correlation_Co-efficient":x}, index=[0])
                cdf = pd.concat([new_row,cdf.loc[:]]).reset_index(drop=True)
    
    return(cdf)


corr_df = find_correlation(df)
display(corr_df.head())
```
